The datasets has different names for target variables according to kaggle. In the next cell we will rename de the outcome/quality to 'target' just to keep all of them equal

```python
import pandas as pd
import numpy as np
from datetime import timedelta, date
import datetime as dt


diabetes.rename({'Outcome' : 'target'}, inplace=True, axis=1)

wine['target'] = np.select(
    [
        wine['quality'] == 'bad' , 
        wine['quality'] == 'good'
    ], 
    [
        0, 
        1
    ]
)

# removing the old target
wine = wine[wine.columns.drop(list(wine.filter(regex='quality')))]
```
